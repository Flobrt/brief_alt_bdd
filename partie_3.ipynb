# PARTIE SQLALCHEMY / MYSQL

```python
import sqlalchemy as db
import pandas as pd
```

```python
engine = db.create_engine(f"mysql+pymysql://root:password@localhost:40000/brief2_csv")
```

```python
def requete(engine, rqt):
    result = engine.execute(db.text(rqt))
     # Transformer le résultat en liste de dictionnaires (pour pandas)
    data = result.fetchall()
    # Récupérer les noms des colonnes
    columns = result.keys()
    # Créer un DataFrame pandas
    df = pd.DataFrame(data, columns=columns)
    return df
```

```python
rqt = '''
    SELECT * 
    FROM clients 
    LIMIT 10;
'''

requete(engine, rqt).head()
```

```python
rqt ='''
    SELECT * 
    FROM clients
    LEFT JOIN ventes ON clients.client_id = ventes.client_id
    LEFT JOIN produits ON ventes.product_id = produits.product_id
    LIMIT 10;
'''

requete(engine, rqt).head()
```

```python
rqt = '''
    SELECT * 
    FROM ventes 
    LEFT JOIN produits ON ventes.product_id = produits.product_id
    LEFT JOIN clients ON ventes.client_id = clients.client_id
    LIMIT 10;
'''

requete(engine, rqt).head()
```
