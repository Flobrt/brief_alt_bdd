```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

## Cas pratique n°1 

```python
site = "https://www.basketball-reference.com/leagues/NBA_2022_per_game.html"

response = requests.get(site)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
players = []
excluded_class = 'thead'

# Recupération des données 
for row in soup.find('tbody').find_all('tr')[1:-1]:
    if excluded_class not in row.get('class', []):
        player = []
        player.append(row.find('th').text)
        for j in row.find_all('td'):
            player.append(j.text)
        players.append(player)
```

```python
# Recupération des colonnes
colonnes = [i.text for i in soup.find_all('tr')[0].find_all('th')]
```

```python
# Création du DataFrame
df = pd.DataFrame(players, columns=colonnes)
df.drop(columns=['Awards'], inplace=True)
df.head()
```

```python
df.to_csv('data/cas_1/result_partie_1.csv', index=False)
```

## Cas pratique n°2

```python
site = "https://content.codecademy.com/courses/beautifulsoup/cacao/index.html"

response = requests.get(site)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
rows = []

for row in soup.find('table', id='cacaoTable').find_all('tr')[1:]:
    # Recupération des données
    rating = row.find('td', class_='Rating').text
    percent = row.find('td', class_='CocoaPercent').text

    row_data = [rating, percent]
    rows.append(row_data)
```

```python
# Recupération des colonnes
rating = soup.find('table', id='cacaoTable').find('td', class_='Rating').text
percent = soup.find('table', id='cacaoTable').find('td', class_='CocoaPercent').attrs['class'][0]

colonnes = [rating, percent]
```

```python
df = pd.DataFrame(rows, columns=colonnes)
df
```

```python
df.to_csv('data/cas_2/result_partie_2.csv', index=False)
```

```python
df.to_json('data/cas_2/result_partie_2_index.json', orient='index')
df.to_json('data/cas_2/result_partie_2_value.json', orient='values')
df.to_json('data/cas_2/result_partie_2_record.json', orient='records')
```

```python
def get_laptop(site):
    response = requests.get(site)
    soup = BeautifulSoup(response.content, 'html.parser')

    rows = []

    for i in soup.find('div', class_='container test-site').find('div', class_='col-lg-9').find_all('div', class_='product-wrapper card-body'):
        title = i.find('a', class_='title').attrs['title']
        description = i.find('p', class_='description').text
        price = i.find('h4', class_='price').text[1:]

        row = [title, description, price]
        rows.append(row)
    
    colonnes = ['titre', 'description', 'prix']

    df = pd.DataFrame(rows, columns=colonnes)
    return df
```

```python
def get_laptop(site):
    response = requests.get(site)
    soup = BeautifulSoup(response.content, 'html.parser')

    rows = []

    for i in soup.find('div', class_='container test-site').find('div', class_='col-lg-9').find_all('div', class_='product-wrapper card-body'):
        title = i.find('a', class_='title').attrs['title']
        description = i.find('p', class_='description').text
        price = i.find('h4', class_='price').text[1:]
        rating = len(i.find('div', class_='ratings').find_all('span', class_='ws-icon ws-icon-star'))
        row = [title, description, price, rating]
        rows.append(row)
    
    colonnes = ['titre', 'description', 'prix', 'note']

    df = pd.DataFrame(rows, columns=colonnes)
    return df
```

```python
df_response = pd.DataFrame()
for i in range(1,5):
    site = f'https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page={i}'
    df_temp = get_laptop(site)

    # Concat le resultat 
    df_response = pd.concat([df_response, df_temp])
```

```python
df_response['prix'] = df_response['prix'].astype('float')
df_response['note'] = df_response['note'].astype('int')
df_response
```

```python
# 1. Les 3 laptops les mieux notés
df_1 = df_response.nlargest(3, "note") 
df_1
```

```python
# 2. Les 3 laptops les moins bien notés
df_2 = df_response.nsmallest(3, 'note')
df_2
```

```python
# 3. Les 3 laptops les plus chers
df_3 = df_response.nlargest(3, "prix") 
df_3
```

```python
# 3. Les 3 laptops les moins chers
df_4 = df_response.nsmallest(3, "prix") 
df_4
```
